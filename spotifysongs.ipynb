```python
import csv, json

with open("spotify_alltime_top100_songs.csv", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    songs = []
    for row in reader:
        songs.append({
            "rank":        int(row.get("alltime_rank") or 0),
            "title":       row.get("song_title", "").strip(),
            "artist":      row.get("artist", "").strip(),
            "streams":     float(row.get("total_streams_billions") or 0),
            "genre":       row.get("primary_genre", "").strip(),
            "year":        int(row.get("release_year") or 0),
            "energy":      float(row.get("energy") or 0),
            "danceability":float(row.get("danceability") or 0),
            "valence":     float(row.get("valence") or 0),
            "bpm":         float(row.get("bpm") or 0),
        })

js = "const SONGS = " + json.dumps(songs, indent=2) + ";"
with open("data.js", "w") as out:
    out.write(js)

print(f"Done — wrote {len(songs)} songs to data.js")
```
